```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../course_assignments/homework7/star_data.txt",
skiprows=1)
print(data.shape)
#2642 rows
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
print(max(magnitude)) #dimmest star
print(min(magnitude)) #brightest star
print(max(temperature)) #hottest star
print(min(temperature)) #coldest star
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

brightest = np.argmin(log_magnitude)
dimmest = np.argmax(log_magnitude)
hottest = np.argmax(log_temperature)
coldest = np.argmin(log_temperature)

plt.figure(figsize=(8,6))
plt.scatter(log_magnitude, log_temperature, label="Star")

plt.scatter(log_magnitude[brightest], log_temperature[brightest], color='black', marker='*', s=300, label="Brightest Star")
plt.scatter(log_magnitude[dimmest], log_temperature[dimmest], color='purple', marker='*', s=300, label="Dimmest Star")
plt.scatter(log_magnitude[hottest], log_temperature[hottest], color='green', marker='*', s=300, label="Hottest Star")
plt.scatter(log_magnitude[coldest], log_temperature[coldest], color='orange', marker='*', s=300, label="Coldest Star")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.legend(loc='lower left')
plt.title('Hertzsprung-Russel Diagram')
plt.xlabel('Temperature (log K)')
plt.ylabel('Magnitude')

plt.show()
```
